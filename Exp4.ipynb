# Basics of plotting using the matplotlib library in Python


```python
import matplotlib.pyplot as plt
import numpy as np
```

## Basic Line Plot
### Write a Python script to create a simple line plot using matplotlib. Plot the function y=2x+3 for values of x ranging from-10 to 10. Label the x-axis as "X-axis", y-axis as "Y-axis", and give the plot a title "Line Plot of y = 2x + 3"

```python
x = np.arange(-10, 11, 1)
y = 2*x + 3
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Line plot of y = 2x+3")
plt.grid("True")
plt.show()
```

## Multiple Line Plots
###  Create a plot with two lines: `y=x^2` and `y=x^3`, for x ranging from-10 to 10. Use different colors and line styles for each line. Add a legend to differentiate between the two plots

```python
x = np.arange(-10,11,1)
y1 = x**2
y2 = x**3
plt.plot(x,y1, label='y=x^2', color='blue')
plt.plot(x,y2, label='y=x^3', color='red')
plt.legend()
plt.grid(True)
plt.show()
```

## Scatter Plot 
###  Generate a scatter plot using matplotlib with random data. Create two lists of 50 random numbers for x and y coordinates. Use different markers and colors for the points. Label the axes and add a title "Random Scatter Plot".

```python
import random
x = [random.randint(0, 100) for _ in range(50)]
y = [random.randint(0, 100) for _ in range(50)]
plt.scatter(x,y, marker='o', color='red')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

## Bar plot
###  Write a script to create a bar plot that displays the population of five different countries. The countries are "USA", "India", "China", "Brazil", and "Russia". The population values are 331, 1391, 1441, 213, and 146 million, respectively. Label the bars with country names and add a title "Country Population".

```python
countries = ['USA','India','China','Brazil','Russia']
population = [331, 139, 144, 213, 146]
plt.bar(countries, population, color=['red','green','blue','orange','violet'])
plt.grid(True)
plt.show()
```

## Histogram
###  Create a histogram to visualize the distribution of 1000 random numbers generated from a normal distribution (mean = 0, standard deviation = 1). Set the number of bins to 30. Label the axes and title the plot "Histogram of Normally Distributed Data"

```python
data = np.random.normal(0,1,5)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Normally and Distributed Data')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

## Pie Chart
###  Using matplotlib, create a pie chart representing the market share of different companies in a sector. Assume the market shares are: "Company A": 40%, "Company B": 30%, "Company C": 20%, and "Company D": 10%. Add labels, a title "Market Share by Company", and make one of the slices explode.

```python
companies = ['Company A', 'Company B', 'Company C', 'Company D']
shares = [40, 30, 20, 10]
explode = (0.1, 0, 0, 0)
plt.pie(shares, labels=companies, explode = explode, autopct='%1.2f%%', startangle=90)
plt.title('Market Share by Company')
plt.show()
```

## Subplots
###  Create a figure with two subplots: one displaying a sine wave and the other displaying a cosine wave. Plot both functions for x values ranging from 0 to 2π. Ensure that each subplot has its own title and labels.

```python
# Generate x values from 0 to 2π (gives 1000 values)
x = np.linspace(0, 2 * np.pi, 1000)
y_sin = np.sin(x)
y_cos = np.cos(x)
fig,(ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

ax1.plot(x, y_sin, label='Sine Wave')
ax1.set_title("Sine Wave")
ax1.set_xlabel('X-values')
ax1.set_ylabel("sin(x)")
ax1.legend()

ax2.plot(x, y_cos, label='Cosine Wave', color='orange')
ax2.set_title("Cosine Values")
ax2.set_xlabel('X-values')
ax2.set_ylabel('cos(x)')
ax2.legend()

plt.tight_layout()

plt.show()
```

## Customized Plot
###  Create a line plot for the function y=sin (x) over the range 0 to 4π. Customize the plot by changing the line color to red, adding gridlines, and adjusting the x-ticks to show values at 0,π,2π,3π, and 4π with appropriate labels.

```python
x = np.linspace(0, 4*np.pi, 1000)
y = np.sin(x)

plt.plot(x,y,color='red')
xticks = [0, np.pi, 2*np.pi, 3*np.pi, 4*np.pi]    # list for xticks
xtick_labels = ['0', 'π', '2π', '3π', '4π']       # labels for xticks
plt.xticks(xticks, xtick_labels)
plt.title("Customized plot for y=sin(x)")
plt.xlabel('X_values')
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```
