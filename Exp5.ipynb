# Frequency Distributions in Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

###  Question 1: Importing Libraries and Loading Data
 - a. Import the necessary libraries (pandas, numpy, matplotlib, seaborn).
 - b. Load adataset of your choice (e.g., titanic.csv, iris.csv, or any other available dataset).

```python
# url = 'https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv'
titanic = pd.read_csv('Titanic-Dataset.csv')
print(titanic.head())
```

 ### Question 2: Creating a Frequency Distribution Table
 - a. Choose a categorical variable from your dataset.
 - b. Create a frequency distribution table that shows the count of each category

```python
pclass_counts = titanic['Pclass'].value_counts()
print(pclass_counts)
```

```python
pclass_counts = titanic['Pclass'].value_counts().reset_index()
pclass_counts.columns = ['Pclass', 'Count']
print(pclass_counts)
```

### Question 3: Visualizing the Frequency Distribution with a Bar Plot
 - a. Create a bar plot to visualize the frequency distribution of the categorical variable you selected in Question 2

```python
# plt.figure(figsize=(8, 6))
plt.bar(pclass_counts.Pclass, pclass_counts.Count, color=['r', 'g', 'b'])
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Frequency Distribution of Passenger Class')
plt.show()
```

 ## Question 4: Analyzing Continuous Data with a Histogram
- a. Choose a continuous variable from your dataset and create a histogram to visualize its frequency distribution

```python
# Histogram for 'Age'
plt.figure(figsize=(8, 6))
plt.hist(titanic['Age'].dropna(), bins=30, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

## Question 5: Using Seaborns distplot
- a. Create a distribution plot using Seaborn's distplot for the continuous variable selected in Question 4. Include a kernel density estimate (KDE)

```python
# Distribution plot for 'Age' with KDE
plt.figure(figsize=(8, 6))
sns.histplot(titanic['Age'].dropna(), bins=30, kde=True)
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Distribution Plot of Age with KDE')
plt.show()
```

##  Question 6: Cumulative Frequency Distribution
 - a. Create a cumulative frequency distribution plot for the continuous variable selected in Question 4.

```python
# Cumulative frequency distribution plot for 'Age'
plt.figure(figsize=(8, 6))
sns.histplot(titanic['Age'].dropna(), bins=30, cumulative=True, kde=False)
plt.xlabel('Age')
plt.ylabel('Cumulative Frequency')
plt.title('Cumulative Frequency Distribution of Age')
plt.show()
```

 ## Question 7: Grouped Frequency Distribution
 - a. Create a grouped frequency distribution table for a continuous variable by dividing it into intervals (bins)

```python
# Grouped frequency distribution for 'Age'
age_bins = pd.cut(titanic['Age'].dropna(), bins=10)
age_grouped = age_bins.value_counts().reset_index()
age_grouped.columns = ['Age Group', 'Count']
print(age_grouped)
```

 ## Question 8: Comparing Frequency Distributions
 - a. Select two categorical variables from your dataset and create a side-by-side bar-plot to compare their frequency distributions

```python
# Side-by-side bar plot for 'Pclass' and 'Sex'
# plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Sex', data=titanic)
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Frequency Distribution of Passenger Class by Sex')
plt.show()
```
